```python
## theoretical expressions : probability density and first passage time, see Equation 29

import scipy.integrate as sp_int
import scipy.special as sp_spec
def Psi(Array):
    return np.exp(Array*Array)*(sp_spec.erf(Array)+1)

## First passage time see. e.g. Brunel & Hakim, 1998
def Brunel_function(nu, J, tau, El, Vthre, Vreset):
    x1=(Vreset-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    x2=(Vthre-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    integral = sp_int.quad(Psi,x1,x2)
    if integral[0]<float('inf') and integral[0]>0:
        T=Tm*np.sqrt(np.pi)*integral[0]
        freq=1.0/T
    else:
        freq=1e-9 # arbitrary for "negligible"
    return freq
    
def Proba_distrib(v, nu, J, tau, El, Vthre, Vreset):
    def rescale(vvv):
        return (vvv-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    def exp_square(y):
        return np.exp(y**2)
    nu_out = Brunel_function(nu, J, tau, El, Vthre, Vreset)
    low_bound = np.amax([v,Vreset])
    #print rescale(v), rescale(low_bound)
    #print low_bound, rescale(low_bound), float(rescale(Vthre))
    integral = sp_int.quad(exp_square, rescale(low_bound), rescale(Vthre))
    return 2*nu_out/J/np.sqrt(tau*nu)*np.exp(-rescale(v)**2)*integral[0]
 
```

```python
Tm = 1. # 20ms
DV = 0.01 # 0.1 mV
vthre = 1.
vrest = 0.
vv = np.linspace(-.1, 1.,5e2)

pp0 = np.zeros(vv.size)
pp1 = np.zeros(vv.size)
pp2 = np.zeros(vv.size)
for iv in range(vv.size):
    pp0[iv] = Proba_distrib(vv[iv], 70., DV, Tm, vrest, vthre, vrest) 
    pp1[iv] = Proba_distrib(vv[iv], 85., DV, Tm, vrest, vthre, vrest) 
    pp2[iv] = Proba_distrib(vv[iv], 95., DV, Tm, vrest, vthre, vrest) 
nu0 = Brunel_function(70., DV, Tm, vrest, vthre, vrest)
nu1 = Brunel_function(85., DV, Tm, vrest, vthre, vrest)
nu2 = Brunel_function(95., DV, Tm, vrest, vthre, vrest)
```

```python
fig = plt.figure(figsize=(6,4))
plt.plot(vv, pp0, 'k', lw=4, alpha=.6, label=r"$\nu_{in} .\tau=70.$")
plt.plot(vv, pp1, 'b', lw=4, label=r"$\nu_{in}.\tau=85.$")
plt.plot(vv, pp2, 'r', lw=4, label=r"$\nu_{in}.\tau=95.$")
plt.xlabel(r"membrane potential ")
plt.ylabel(r"probability density")
plt.legend(frameon=False, loc='upper left',prop={'size':17})#,bbox_to_anchor = (0.1, 0.92))
plt.twinx()
plt.plot([-.2,0.,0.,1.],[0,0,nu0,nu0],'k--',lw=1.5)
plt.plot([-.2,0.,0.,1.],[0,0,nu1,nu1],'b--',lw=1.5)
plt.plot([-.2,0.,0.,1.],[0,0,nu2,nu2],'r--',lw=1.5)
plt.ylim([0,.5])
plt.ylabel(r"probability flux")
plt.xlim([-.1,1.05])
plt.xticks([0,.2,.4,.6,.8,1.],[r" $V_{reset}$",'0.2','0.4','0.6','0.8',r"$V_{thre}$"]);
```

```python
## Numerical simulations of the excitatory delta-synapse case

def SingleSimulation(nu, v_rest, v_thre, DV, Tm): # one simulation for a given frequency of input
    pre_spk_train = np.random.exponential(1./nu, 1e5)

    counter_spk = 0 # spike counter
    v = v_rest # membrane potential at last pre-syn event
    t = pre_spk_train[0] # time at last pre-syn event
    for pre_spk in pre_spk_train[1:]:
        # we start by testing the effect of the last spike
        # did this last event could trigger a spike ??
        v+=DV#/Gl
        if v>v_thre:
            # it triggers a spike
            v = v_rest
            counter_spk +=1
              
        v = v_rest+(v-v_rest)*np.exp(-pre_spk/Tm)
        t+=pre_spk   
    return counter_spk/t
```

```python
# to get the figure of Brunel & Sergi 1998, we take their values
Tm = 1. # 20ms
DV = 0.01 # 0.1 mV
vthre = 1
vrest = 0
nu, fr = np.array([(_, SingleSimulation(_, vrest, vthre, DV, Tm)) for _ in np.arange(70,102.5,2.5)/Tm]).T
nu_th, fr_th = np.array([(_, Brunel_function(_, DV, Tm, vrest, vthre, vrest)) for _ in np.linspace(70,100)/Tm]).T
```

```python
fig = plt.figure(figsize=(4.7,3))
plt.xlabel(r"$\nu_{in}\,.\,\tau$");plt.ylabel(r"$\nu_{out}\,.\,\tau$")
plt.ylim([-.02,.35]);plt.yticks(np.arange(0,.35,.1))
plt.plot(nu, fr, 'kD', label='numerical sim.')
plt.plot(nu_th, fr_th, 'k-', label='diffusion approx.')
#plt.tight_layout()
plt.legend(loc=2,frameon=False,prop={'size':14})
```
