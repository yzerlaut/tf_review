the problem is defined as :

$$
\tau_I \frac{dI}{dt} = -I(t) + J . S(t)
$$

$$
C_m \frac{dV}{dt} = g_L (E_L - V(t)) + I(t)
$$

from the equation for $I(t)$, we get:

$$
I(t) = \frac{J}{\tau_I}  \sum_{\{ t_k \}} e^{\frac{t_k -t }{\tau_I}} H(t-t_k)
$$

posons $J_V = J_I / C_m$ and we shift $V$ by $E_L$, we obtain :

$$
\tau_m \frac{dV}{dt} = - V + J_V \frac{\tau_m}{\tau_I}  \sum_{\{ t_k\}} e^{\frac{t_k -t }{\tau_I}} H(t-t_k)
$$

Posons :

$$
\tau_m \frac{dV}{dt} = - V + \frac{\tau_m}{\tau_I} \alpha(t)
$$

we consider the effect of the arrival at $t_1$ of an event 
(the one before was at $t_0$, the one after will be at $t_2$):

$\alpha(t_1^+) = \alpha_1 = J_V \sum_{t_k \leq t_1 }  e^{\frac{t_k -t_1 }{\tau_I}}$ i.e. after the $J_V$ increment at $t_1$

$\alpha(t_1 < t < t_2) = \alpha_1  e^{-\frac{t-t_1}{\tau_I}}$

$V(t_1)=V_1$ (continuous function of $t$)

then for $t_1 < t < t_2$:

$$
\tau_m \frac{dV}{dt} = - V + \frac{\tau_m}{\tau_I} \alpha_1 e^{\frac{t_1 -t }{\tau_I}}
$$

the solution is:

$$
V(t>t_1) = (V_1 + \frac{\tau_m}{\tau_m - \tau_I} \alpha_1)e^{-\frac{t-t_1}{\tau_m}} - \frac{\tau_m}{\tau_m - \tau_I}  \alpha_1 
e^{-\frac{t-t_1}{\tau_I}}
$$

```python
def PSP(t, Tm, Ts, I0, V0=0):
    """
    t is actually t-t_presyn_spike
    """
    if Ts==0:
       return (V0+I0)*np.exp(-t/Tm) 
    if Tm!=Ts:
        return (V0+I0*Tm/(Tm-Ts))*np.exp(-t/Tm)-I0*Tm/(Tm-Ts)*np.exp(-t/Ts)
    else:
        return (V0+I0*t/Tm) * np.exp(-t/Tm)

t = np.linspace(0,50,1e2)
fig = plt.figure()
t2 = np.linspace(-5,0,5)
plt.plot(np.concatenate([t2,t]), np.concatenate([0*t2, PSP(t, 10, 0, 0.1)]), 'k', alpha=.8, lw=3, label=r"$\tau_S$ = 0 ms")
plt.plot(np.concatenate([t2,t]), np.concatenate([0*t2, PSP(t, 10, 5., 0.1)]), 'b', lw=3, label=r"$\tau_S$ = 5 ms")
plt.plot(np.concatenate([t2,t]), np.concatenate([0*t2, PSP(t, 10, 10., 0.1)]), 'r', lw=3, label=r"$\tau_S$ = 10 ms")
plt.xlabel('time from synaptic event (ms)')
plt.ylabel('membrane potential (mV)')
plt.legend()
#fig.savefig('/home/yann/files/work/transfer_functions/review_previous_work/f2_1.pdf')
```

for the event based driven strategy, we need to derive the maximum of this PSP time course

$$
t_{max} = \frac{\tau_m \tau_I}{\tau_m - \tau_I} 
Ln \Big( \frac{\tau_m^2 \alpha_1}{\tau_I (\tau_m - \tau_I) (V_1 + \frac{\tau_m \alpha_1}{\tau_m - \tau_I}  ) } \Big)
$$

Now in the case $V(t_{max})>V_{thre}$, there will be a spike, we need to determine its precise timing, this corresponds to finding the roots of :

$$
V_{thre} = (V_1 + \frac{\tau_m}{\tau_m - \tau_I} \alpha_1)e^{-\frac{t-t_1}{\tau_m}} - \frac{\tau_m}{\tau_m - \tau_I}  \alpha_1 
e^{-\frac{t-t_1}{\tau_I}}
$$

we solve this with a newton algorithm !


```python
# we prepare the numerical functions with some functions used in the algorithm

def Tmax(Tm, Ts, I0, V0=0.): # delay of the max jump value after a synaptic event
    if Ts==0:
        Tmax=0
    elif Tm!=Ts: 
        Tmax = Tm*Ts/(Tm-Ts) * np.log( Tm**2*I0 / ( Ts*(Tm-Ts)*(V0+Tm*I0/(Tm-Ts)) ) )
    else:
        Tmax = Tm*(1-V0/I0)
    return Tmax

from scipy.optimize import newton


def find_threshold_crossing(Vthre, Tm, Ts, I0, V0=0.):
    """ newton(func, x0, fprime=None, args=(), tol=1.48e-08, maxiter=50,  """
    # to be executed only if there is an actual threshold crossing V(tmax)>Vth
    if Ts==0:
        res=0 # in this case, the spike is instantaneous
    elif Tm!=Ts:
        fun = lambda x:  (V0+I0*Tm/(Tm-Ts))*np.exp(-x/Tm)-I0*Tm/(Tm-Ts)*np.exp(-x/Ts) -Vthre
        res = newton(fun, 0.)#, method='SLSQP', bounds=bnds)
    else:
        fun = lambda x:  (V0+I0*x/Tm) * np.exp(-x/Tm) -Vthre
        res = newton(fun, 0.)#, method='SLSQP', bounds=bnds)
    return res
    
J = 0.1
Tm = 10.
Ts1 = 5.
v0 = 0.1
threshold = J/5.+v0

t = np.linspace(0,20,1e2)
plt.figure(figsize=(10,5))
plt.title('Constructing the algorithm for the Event-based simulations')
plt.plot(t, PSP(t, Tm, Ts1, v0+J, V0=v0), 'r', label=r"$\tau_S$ = 5 ms")
plt.plot(t, PSP(t, Tm, Tm, v0+J, V0=v0), 'b', label=r"$\tau_S$ = 10 ms")
tmax, vmax = Tmax(Tm, Ts1, v0+J, V0=v0), PSP(Tmax(Tm, Ts1, v0+J, V0=v0), Tm, Ts1, v0+J, V0=v0)
plt.plot([0, tmax, tmax], [vmax, vmax, 0],'r--', lw=2, label=r"$v_{max},\,\, t_{max}$")
tmax, vmax = Tmax(Tm, Tm, v0+J, V0=v0), PSP(Tmax(Tm, Tm, v0+J, V0=v0), Tm, Tm, v0+J, V0=v0)
plt.plot([0, tmax, tmax], [vmax, vmax, 0],'b--', lw=2)
tthre = find_threshold_crossing(threshold, Tm, Ts1, v0+J, V0=v0)
plt.plot([0, tthre, tthre], [threshold, threshold, 0],'r-', lw=2, label=r"$v_{thre}, \,\, t_{thre}$")
tthre = find_threshold_crossing(threshold, Tm, Tm, v0+J, V0=v0)
plt.plot([0, tthre, tthre], [threshold, threshold, 0],'b-', lw=2)
plt.plot(t, threshold*np.ones(t.size),'k-',lw=4, alpha=.4, label='threshold')
plt.xlabel('time from synaptic event')
plt.ylabel('membrane potential')
plt.legend()
```

```python
## Now we have all the quantities need for the numerical simulations

## Numerical simulations of Brunel and Sergi
# now that we have all the quantities, we can run the simulation


vvec = []

def Shotnoise_Sim(nu, J, Tm, Ts, v_thre, v_reset, max_event=1e5):
    if nu==0: print "NON ZERO FREQUENCY NEEDED!!"
    
    # we prepare the pre-synaptic spike train
    pre_spk_train = np.random.exponential(1./nu, max_event)
    
    counter_spk = 0 # spike counter
    i, v = 0., v_reset # initial conditions
    t_last = pre_spk_train[0] # time at last pre-syn event
    for pre_spk in pre_spk_train[1:]:
        # we start by testing the effect of the last spike
        i+=J # there has been an increment of the input
        # did this last event could trigger a spike ??
        if PSP(Tmax(Tm, Ts, i, V0=v), Tm, Ts, i, V0=v)>v_thre:
            # if it can, when  would it trigger it ?
            tspk = find_threshold_crossing(v_thre, Tm, Ts, i, V0=v)
        else:    
            tspk = np.inf # this means no post-syn spike !
        # so now, we test if there is no pre-syn spike, that happens before the post-syn spike
        if tspk<pre_spk: # then SPIKE
            counter_spk +=1
            v = v_reset # we reset the potential
            #print tspk
        else:
            v = PSP(pre_spk, Tm, Ts, i, V0=v)  # we update the potential
        # then, whatever happens, we update the time to the next spike, as nothing can happen without any event
        # (no leak because El = v_reset)
        if Ts!=0:
            i = i*np.exp(-pre_spk/Ts)
        else:
            i=0
        t_last += pre_spk
    return counter_spk/np.cumsum(pre_spk_train)[-1]
        
       
#Shotnoise_Sim(nu, J, Tm, Ts, v_thre, v_reset, max_event=10)
Shotnoise_Sim(100., 0.01, 1., 0.3, 1., 0., max_event=1e5)   
```

# Analytical expression

```python
import scipy.integrate as sp_int
import scipy.special as sp_spec
def Psi(Array):
    return np.exp(Array*Array)*(sp_spec.erf(Array)+1)

## First passage time see. e.g. Brunel & Hakim, 1998
def Brunel_function(nu, J, tau, El, Vthre, Vreset):
    x1=(Vreset-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    x2=(Vthre-El-nu*J*tau)/(J*np.sqrt(nu*tau))
    integral = sp_int.quad(Psi,x1,x2)
    if integral[0]<float('inf') and integral[0]>0:
        T=Tm*np.sqrt(np.pi)*integral[0]
        freq=1.0/T
    else:
        freq=1e-9 # arbitrary for "negligible"
    return freq

## now with correction for correlated noise input 
# (membrane equation with colored noise input, correlation time : Tau )
def BrunelFunc_colored(nu, J, tau, El, Vthre, Vreset, Ts):
    # taken from Brunel and Sergi 1998 :
    mu = J*nu*tau
    s = J*np.sqrt(nu*tau)
    theta_hat = np.sqrt(2)*(Vthre-mu)/s
    theta = Vthre + s*( 1.03*np.sqrt(Ts/Tm) - Ts/Tm*(-0.34 + 0.29*theta_hat) )
    return Brunel_function(nu, J, tau, El, theta, Vreset)
```

```python
from scipy import special, integrate

def gauss_integral(x, mu, sigma):
    """integral of normal distribution from x to inf"""
    return special.erfc((x-mu)/np.sqrt(2)/sigma)/2.

def gaussian_1d(x, mu, sigma):
    return 1./(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2.*sigma**2))

def erf_integral(x, mu, sigma):
    """integral from -inf to x of gauss_integral"""
    x_ = (x-mu)/sigma/np.sqrt(2)
    return (x_*(1+special.erf(x_))+np.exp(-x_**2)/np.sqrt(np.pi))*sigma*np.sqrt(2)

def ou_rate_markov_anal(mu, sigma, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    correction = np.sqrt((tau+tau_m)/tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold), sigma)/2.
    prob = markov_integral*stationary/correction
    return prob

def ou_rate_markov_anal2(mu, sigma, tau, threshold):
    sigma_markov = np.sqrt(sigma**2/(1./tau_m*(1./tau_m+1./tau)))
    mu_markov = (mu*tau_m)
    
    stationary = gaussian_1d(threshold, mu_markov, sigma_markov)
    markov_integral = erf_integral(0, (threshold-mu_markov)/np.sqrt(tau), sigma_markov/tau)/2.
    prob = markov_integral*stationary
    return prob
```

## Brunel Functions

```python
import scipy.integrate as sp_int
import scipy.special as sp_spec
def Psi(Array):
    return np.exp(Array*Array)*(sp_spec.erf(Array)+1)

## First passage time of an Ornstein Uhlenbeck process
# (membrane equation with white noise input)
## without correction because of the colored noise we would get
def BrunelFunc(mVm, sVm, Tm, Vthre, Vreset):
    x1=(Vreset-mVm)/(sVm)
    x2=(Vthre-mVm)/(sVm)
    integral = sp_int.quad(Psi,x1,x2)
    if integral[0]<float('inf') and integral[0]>0:
        T=Tm*np.sqrt(np.pi)*integral[0]
        freq=1.0/T
    else:
        freq=1e-9
    return freq
## now with correction for correlated noise input 
# (membrane equation with colored noise input, correlation time : Tau )
def BrunelFunc_colored(mVm, sVm, Tm, Vthre, Vreset, Tau):
    # taken from Brunel and Sergi 1998 :
    return BrunelFunc(mVm, sVm, Tm, Vthre + 1.03*np.sqrt(Tau/Tm)*sVm, Vreset)

sVm = np.sqrt(sigmas**2/(1./tau_m*(1./tau_m+1./tau)))
brunel0 = np.array([(BrunelFunc(mVm, _, tau_m, threshold, reset))
                      for _ in sVm]).T
brunel1 = np.array([(BrunelFunc_colored(mVm, _, tau_m, threshold, reset, tau))
                      for _ in sVm]).T
```
